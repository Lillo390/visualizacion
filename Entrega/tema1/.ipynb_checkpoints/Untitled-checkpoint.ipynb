```python
from sklearn.gaussian_process import GaussianProcessRegressor
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Compute the Gaussian process fit
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit_1, MSE_1 = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit_1 = 2 * np.sqrt(MSE)  # 2*sigma ~ 95% confidence region
```

```python
# Visualize the result
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(xfit, yfit_1 - dyfit_1, yfit_1 + dyfit_1,
                 color='gray', alpha=0.2)
plt.xlim(0, 10);
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error


# define the models and draw some data
model= lambda x: x * np.sin(x)

xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Compute the Gaussian process fits
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

rf = RandomForestRegressor()
rf.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit_1, MSE_1 = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit_1 = 2 * np.sqrt(MSE_1)  # 2*sigma ~ 95% confidence region

yfit_2 = rf.predict(xfit[:, np.newaxis])
MSE_2 = mean_squared_error(model(xfit), yfit_2, multioutput='raw_values')
dyfit_2 = 2 * np.sqrt(MSE_2)


```

```python
# Visualize the results
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit_1, '-', color='gray')
plt.fill_between(xfit, yfit_1 - dyfit_1, yfit_1 + dyfit_1,
                 color='gray', alpha=0.2)

plt.plot(xfit, yfit_2, '--', color='blue')
plt.fill_between(xfit, yfit_2 - dyfit_2, yfit_2 + dyfit_2,
                 color='blue', alpha=0.2)

plt.xlim(0, 10)
plt.ylim(-10, 20)
plt.show()
```
